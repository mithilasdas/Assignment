**3)Who scores the most on average for math, reading and writing based on**

**1.**   **gender**
**2.**   **Test preparation course**



```python
total=data['math score']+ data['reading score'] + data['writing score']
```

```python
data['average']=total/3
```

```python
data
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.barplot(x=data['gender'],y=data['average'])
plt.subplot(1,2,2)
sns.barplot(x=data['test preparation course'],y=data['average'])
```

**Inference**

1)***female scores more than male based on gender***

2)***those who completed the test preparation course scored more***


**4. What do you think about the scoring variation for math, reading and writing based on**

**1.** **gender**
**2.** **test preparation course** 






```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.barplot(x=data['gender'],y=data['math score'])
plt.subplot(1,3,2)
sns.barplot(x=data['gender'],y=data['reading score'])
plt.subplot(1,3,3)
sns.barplot(x=data['gender'],y=data['writing score'])
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
sns.barplot(x=data['test preparation course'],y=data['math score'])
plt.subplot(1,3,2)
sns.barplot(x=data['test preparation course'],y=data['reading score'])
plt.subplot(1,3,3)
sns.barplot(x=data['test preparation course'],y=data['writing score'])
```

**inference:**
>***based on gender***
1.  ***male candidates scored more in maths test***
2.  ***female candidates scored more in reading test***
3.  ***female candidates scored more in writing test***
> ***based on test preparation course***
1.   ***those candidates who completed the test preparation course scored more in all the tests.***
