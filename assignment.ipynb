# **5. Perform meaningful visualizations using the dataset**

```python

data['Classification'].value_counts().plot(kind='pie')
plt.title('classification')
```

```python
num_data=data[['SW','SL','PL','PW']]
```

```python
sns.pairplot(num_data)
plt.title('relation between SW,SL,PL and PW')
```

```python
sns.barplot(x=data['Classification'],y=data['PL'])
plt.title('petal length of different classification')
```

```python
sns.barplot(x=data['Classification'],y=data['SL'])
plt.title('Sepal length of different classification')
```

```python
sns.barplot(x=data['Classification'],y=data['PW'])
plt.title('petal width of different classification')
```

```python
sns.barplot(x=data['Classification'],y=data['SW'])
plt.title('sepal width of different classification')
```

```python
sns.boxplot(x=data['Classification'],y=data['PL'])
plt.title('classification vs petal length')
```

```python
sns.violinplot(x=data['Classification'],y=data['PW'])
plt.title('classification vs petal width')
```

```python
data['Classification'].value_counts().plot(kind='bar')
plt.title('count of classification')
```

```python
sns.countplot(x=data['Classification'])
```

```python
sns.scatterplot(x=data['SW'],y=data['SL'],hue=data['Classification'])
plt.title('scatterplot between sepal width and sepal length')
```
